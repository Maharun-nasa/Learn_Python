## Python Loop :
- While Loop
- For Loop

````python
# Creating a while loop
i = 1
while i<5: # here i is the variable , while loop will run until i is less than 5
    print(i)
    i+=1 # i+=1 means i=i+1 , so the value of i will increase by 1 in each iteration
else:
    print('i is no longer less than 5')

# Creating a for loop 
fruits = ['apple','mango','jackfruit']
for x in fruits : # x is the variable here where the items of the list will be stored one by one
    print(x)

# for loop in the range function range()
for a in range (2,7): # here 2 is starting value and 7 is ending value
    print(a)

for b in range (5,50,15): # here 5 is starting value , 50 is ending value ,15 is the step count
    print(b)
    
# Looping through a string
for x in 'mango': # here x is the variable where each character of the string will be stored one by one
    print(x) 

# Else in for loop in the range function
for y in range(6):
    print(y)
else :
    print("Finished")

# Nested loops 
shade = ['light','deep']
color = ['red','blue','green']

for z in shade: # z is the variable here where the items of the list shade will be stored one by one
    for w in color: # w is the variable here where the items of the list color will be stored one by one
        print(z,w) # this will print the combination of shade and color that's why nested loops are used

````

- The continue Statement:

With the continue statement we can stop the current iteration of the loop, and continue with the next

````python
# Use of continue statement In for loop
fruits = ['apple','mango','jackfruit','orange']
for x in fruits:
    if x == 'jackfruit': # when x is equal to jackfruit the loop will skip this item and continue the loop
        continue
    print(x)

# Use of continue statement In while loop
i = 0
while i<5:
    i += 1
    if i == 3 : # when i is equal to 3 the loop will skip this item and continue the loop
        continue
    print(i)
````

- The Break Statement:

````python
# break Statement in while Loop:
i = 1
while i<5: # here i is the variable , while loop will run until i is less than 5
    print(i)
    if i == 3 : # when i is equal to 3 the loop will stop executing for using break statement  
        break
    i += 1

# break Statement in For Loop:
fruits = ['apple','mango','jackfruit','orange']
for x in fruits:
    print(x)
    if x == 'jackfruit':  # when x is equal to jackfruit the loop will stop executing
        break
    
# Exit the loop when x is "jackfruit", but this time the break comes before the print:
fruits = ['apple','mango','jackfruit','orange']
for y in fruits:
    if y == 'jackfruit':
        break
    print(y)
````

## Parameters or Arguments ?

### From a function's perspective :

## Arguments, *args
```
If you do not know how many arguments that will be passed into your function, add a (*) before the parameter name in the function definition.
```
This way the function will receive a tuple of arguments, and can access the items accordingly :

````python
def my_function(*city):
    print("I live in " + city[1]) # here city will call the second item of the argument because index of 1 means second item

my_function('rajshahi','dinajpur','dhaka')
````

## Keyword Arguments
```
You can also send arguments with the key = value syntax.
```
This way the order of the arguments does not matter.

````python
def my_function(city1,city2,city3):
    print("Most populated city in BD is : " + city3)

my_function(city1='dinajpur',city2='rajshahi',city3='dhaka')
````

## Arbitrary Keyword Arguments, **kwargs
```
If you do not know how many keyword arguments that will be passed into your function, add two asterisk: ** before 
the parameter name in the function definition.
```
This way the function will receive a dictionary of arguments, and can access the items accordingly :

````python
def my_function(**lname):
    print("Her last name is " + lname['lastname'])

my_function(firstname = 'maharun',lastname = 'nishi')

# Passing a List as an Argument
def function(fruits):
    for x in fruits:
        print(x)

fruits = ['apple','banana','cherry']
function(fruits)
````

## Python Inheritance :
Inheritance allows us to define a class that inherits all the methods and properties from another class.
- Parent class is the class being inherited from, also called base class.
- Child class is the class that inherits from another class, also called derived class.
  
Create a Parent Class : Any class can be a parent class, so the syntax is the same as creating any other class

````python
class Person:
    def __init__(self, fname, lname):
        self.firstname = fname
        self.lastname = lname
    def printname(self):
        print(self.firstname, self.lastname)
    
x = Person("Maharun", "Nishi")
x.printname()
````

````python
# Create a Child Class
class Person:
    def __init__(self, fname, lname):
        self.firstname = fname
        self.lastname = lname
    def print_name(self):
        print(self.firstname, self.lastname)
class child_class(Person):
    pass
y= child_class("Maharun", "Nishi")
y.print_name()
````

## Exception / Try Except :
- The ```try``` block lets you test a block of code for errors.
- The `except` block lets you handle the error.
- The `else` block lets you execute code when there is no error.
- The `finally` block lets you execute code, regardless of the result of the try- and except 
blocks.

````python
# The try block will generate an exception, because x is not defined:
try:
    print(x)
except:
    print("An exception occurred")
````

````python
# Print one message if the try block raises a NameError and another for other errors:
try:
    print(x)
except NameError:
    print("Variable x is not defined")
except:
    print("Something else went wrong")

# In this example, the try block does not generate any error:
try:
    print("Hello World")
except:
    print("Something went wrong")
# The else block will execute if the try block does not raise an error:
else:
    print("Nothing went wrong")
````

````python
try:
    print(x)
except:
    print('something went wrong')
finally:
    print('The "try except" is finished')
````
